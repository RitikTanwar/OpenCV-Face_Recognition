```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
img=cv2.imread('Scrapy/amazon/amazon_data/watch/Vills Laurrens Pack of 4 Multicolour Analog Analog Watch for Men and Boy/Main_image.jpg')           
```

```python
plt.imshow(img)
plt.show()
```

Color of image is different from original ones because openCv read images in BGR not RGB bydefault so we have to swap them. This is provided by openCv already

```python
newImg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(newImg)
plt.show()
```
